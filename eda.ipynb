```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
df = pd.read_csv(r'Projeto Base\data\penguins.csv')
df.head()
```

```python
dfg = (
    df
    .groupby(['island', 'species'])
    ['body_mass_g'].mean()
    .reset_index()
    .round(2)
    .rename(columns={
        'island': 'Ilha'
        , 'species': 'Espécies'
        , 'body_mass_g': 'Massa (g)'
    })
)

dfg.head()
```

```python
fig = (
    px.bar(
        dfg
        , x="Ilha"
        , y="Massa (g)"
        , color='Espécies'
        , title="Peso Médio dos Pinguins por ilha"
        , barmode="group"
    )
)

fig.show()
```
